### ROC and EER on voxceleb

```python
import os
import torch
import numpy as np
from datasets import load_dataset
from transformers import Wav2Vec2FeatureExtractor, WavLMForXVector
from sklearn.metrics import roc_curve, roc_auc_score

# Load dataset and model
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained('microsoft/wavlm-base-sv')
model = WavLMForXVector.from_pretrained('microsoft/wavlm-base-sv')
```

```python
# Read list_test_hard.txt and parse rows
file_path = '/content/voxceleb/VoxCeleb1/list_test_hard.txt'
with open(file_path, 'r') as file:
    rows = file.readlines()

# Extract audio paths from rows
audio_paths = []
for row in rows[:5]:  # Considering only the first 10 rows
    _, path1, path2 = row.split()
    audio_paths.append((os.path.join('/content/voxceleb/VoxCeleb1/dev', path1), os.path.join('/content/voxceleb/VoxCeleb1/dev', path2)))
```

```python
print(audio_paths)
```

```python
similarities = []
# Load audio files and process them through the model to obtain embeddings
for path1, path2 in audio_paths:
    audio_input1, _ = torchaudio.load(path1)
    audio_input2, _ = torchaudio.load(path2)

    audio_inputs = [audio_input1.squeeze().numpy(), audio_input2.squeeze().numpy()]
    inputs = feature_extractor(audio_inputs, padding=True, return_tensors="pt")

    embeddings = model(**inputs).embeddings
    embeddings = torch.nn.functional.normalize(embeddings, dim=-1).cpu()
    cosine_sim = torch.nn.CosineSimilarity(dim=-1)
    similarity = cosine_sim(embeddings[0], embeddings[1])
    similarities.append(similarity)
    #embedding = torch.nn.functional.normalize(embedding, dim=-1).cpu().detach().numpy()
    #embeddings.append(embedding)
```

```python
print(similarities)
```

```python
import torch
from sklearn.metrics import roc_curve, roc_auc_score

# Extract similarity values from tensors
similarities = [sim.item() for sim in similarities]

# Prepare labels (0 for same speaker, 1 for different speaker)
labels = [0 if i % 2 == 0 else 1 for i in range(len(similarities))]

# Calculate ROC curve and AUC
fpr, tpr, _ = roc_curve(labels, similarities)
roc_auc = roc_auc_score(labels, similarities)

# Plot ROC curve
import matplotlib.pyplot as plt
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Calculate Equal Error Rate (EER)
eer = np.min(np.abs(tpr - (1 - fpr)))
print(f"Equal Error Rate (EER): {eer:.2%}")
```

```python
import os
import torch
import numpy as np
from datasets import load_dataset
from transformers import Wav2Vec2FeatureExtractor, WavLMForXVector
from sklearn.metrics import roc_curve, roc_auc_score

# Load dataset and model
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained('microsoft/wavlm-base-sv')
model = WavLMForXVector.from_pretrained('microsoft/wavlm-base-sv')
```

```python
import os
import random

# Path to the directory containing the audio files
directory = '/content/katabath_hindi/hindi/valid/audio'

# List to store tuples of audio file pairs
audio_pairs = []

# List to store labels (0 for different speakers, 1 for same speaker)
labels = []

# Variables to count pairs with same speaker IDs and genders
same_ids_and_gender_count = 0

# Traverse through the directory
audio_files = os.listdir(directory)
random.shuffle(audio_files)  # Shuffle the list of audio files
num_files = min(8, len(audio_files))  # Select the minimum of 100 or total files

# Form pairs of audio files
for i in range(0, num_files, 2):
    file1, file2 = audio_files[i:i+2]
    speaker_id1, gender1 = file1.rsplit('-', 2)[1:]  # Extract speaker ID and gender from filename
    speaker_id2, gender2 = file2.rsplit('-', 2)[1:]  # Extract speaker ID and gender from filename
    audio_path1 = os.path.join(directory, file1)
    audio_path2 = os.path.join(directory, file2)
    audio_pairs.append(((audio_path1, speaker_id1, gender1), (audio_path2, speaker_id2, gender2)))

    # Determine label based on speaker IDs and genders
    if speaker_id1 == speaker_id2 and gender1 == gender2:
        labels.append(1)
    else:
        labels.append(0)

# Count the number of pairs with the same speaker IDs and genders
selected_speaker_id, selected_gender = audio_pairs[0][0][1:]  # Get speaker ID and gender of the first pair
for pair in audio_pairs:
    if pair[0][1:] == (selected_speaker_id, selected_gender):
        same_ids_and_gender_count += 1

# Print the count of pairs with the same speaker IDs and genders
print(f"Number of pairs with the same speaker IDs and genders as the first randomly selected pair: {same_ids_and_gender_count}")
```

```python
import torch

similarities = []
# Load audio files and process them through the model to obtain embeddings
for pair in audio_pairs:
    path1, path2 = pair[0][0], pair[1][0]

    # Load audio files
    audio_input1, _ = torchaudio.load(path1)
    audio_input2, _ = torchaudio.load(path2)

    # Extract features and prepare inputs
    audio_inputs = [audio_input1.squeeze().numpy(), audio_input2.squeeze().numpy()]
    inputs = feature_extractor(audio_inputs, padding=True, return_tensors="pt")

    # Obtain embeddings
    embeddings = model(**inputs).embeddings
    embeddings = torch.nn.functional.normalize(embeddings, dim=-1).cpu()

    # Calculate cosine similarity
    cosine_sim = torch.nn.CosineSimilarity(dim=-1)
    similarity = cosine_sim(embeddings[0], embeddings[1])
    similarities.append(similarity.item())
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import numpy as np

# Calculate ROC curve and AUC
fpr, tpr, _ = roc_curve(labels, similarities)
roc_auc = roc_auc_score(labels, similarities)

# Plot ROC curve
import matplotlib.pyplot as plt
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

# Calculate Equal Error Rate (EER)
eer = np.min(np.abs(tpr - (1 - fpr)))
print(f"Equal Error Rate (EER): {eer:.2%}")
```
